# Task 1.2 — Key Assumptions (8 marks)

**Paper:** *Online Discovery and Maintenance of Time Series Motifs* — Mueen et al., KDD 2010

All three assumptions below are specific to the method proposed in this paper, not generic ML assumptions.

---

## Assumption 1

**Assumption:**  
The data stream arrives at a rate λ that is at most 1/(b·m·w), where b is the time cost per floating-point operation, m is the subsequence length, and w is the window size. In other words, the algorithm can process each new data point and complete a full update before the next point arrives.

**Why the method needs it:**  
The algorithm performs O(wm) distance computations on every insertion. If the stream is faster than this computational budget, the algorithm falls behind and cannot maintain the exact motif in real time. The paper's correctness guarantee—that the motif is always exact and up-to-date—collapses if the update cannot finish before the next data point arrives.

**Violation scenario:**  
A high-frequency financial tick stream (e.g., 10,000 samples/second with m=50, w=1000) would generate λ = 10,000 while the budget is 1/(10⁻⁸ × 50 × 1000) = 2000 updates/second—five times faster than the algorithm can handle.

**Paper reference:**  
Section 3 (Notation and Background), final paragraph: "We assume that λ is below this maximum rate until section 6.1, where we remove the restriction." The formula 1/(bmw) appears explicitly in this section.

---

## Assumption 2

**Assumption:**  
Z-normalised Euclidean distance is the correct and sufficient similarity measure for identifying meaningful motifs. That is, two subsequences are considered a motif pair if and only if their z-normalised forms are close in Euclidean space, and this reflects genuine structural repetition in the data.

**Why the method needs it:**  
The entire data structure—N-lists, R-lists, and the optimality of the motif query—is built on Euclidean distance between z-normalised vectors. Observation 2 (the pruning rule for N-lists) and the binary search along the order-line are all derived under this metric. Using a different similarity measure (e.g., DTW or correlation) would invalidate both the pruning logic and the motif guarantee.

**Violation scenario:**  
In ECG analysis, short amplitude plateaus at different voltages may have the same clinical meaning (same morphology). If the clinically meaningful patterns have different shapes but similar amplitude-normalised profiles, z-normalisation may report a spurious subsequence pair as the motif while ignoring the clinically relevant one—especially if the relevant pair has natural amplitude variation.

**Paper reference:**  
Section 2, paragraph after Definition 5: "To make the algorithm invariant to baseline and amplitude scaling, we z-normalize every subsequence." Also Section 2.1 (Why is this problem hard?) explicitly states the algorithm assumes Euclidean distance is used in the metric property.

---

## Assumption 3

**Assumption:**  
The two occurrences of the motif are non-overlapping—that is, their starting indices differ by at least *m* (the subsequence length). The best motif pair is therefore defined as the closest pair among all non-overlapping subsequences.

**Why the method needs it:**  
Without the non-overlapping constraint, the two "closest" subsequences would trivially be adjacent or near-identical windows shifted by a single time step, which share most of their values and are inherently similar regardless of the pattern content. The entire N-list construction enforces non-overlap; a new point is never compared with a subsequence that shares data points (the exclusion condition in Definition 4).

**Violation scenario:**  
A quasi-periodic signal with a pattern whose period is shorter than *m* (e.g., a 20 Hz oscillation captured at m=30 samples) would have many overlapping occurrences within m samples of each other. These would all be excluded by the non-overlapping rule, potentially making the algorithm report a much more distant non-overlapping pair as the motif, even though the true signal structure is a densely repeated waveform at sub-m spacing.

**Paper reference:**  
Definition 4 (Online Motif): "a pair of subsequences (x_{i,m}, x_{j,m}) for 1 ≤ i < i+m ≤ j ≤ t-m+1"—the condition i+m ≤ j explicitly enforces non-overlap. Section 2 also notes: "The reason for considering only the non-overlapping sequences is to avoid trivial matches that are inherently similar because they share most of their values."
