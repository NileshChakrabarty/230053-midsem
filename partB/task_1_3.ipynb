# Task 1.3 — What the Paper Claims to Improve (9 marks)

**Paper:** *Online Discovery and Maintenance of Time Series Motifs* — Mueen et al., KDD 2010

---

## Main Baseline / Prior Method

The primary baseline the paper identifies and compares against is **exact offline batch motif discovery**, specifically the algorithm of **Mueen et al. (SDM 2009)** — "Exact Discovery of Time Series Motifs" [17 in the paper's references]. This is the state-of-the-art offline algorithm that, given a complete time series, finds the exact closest pair of non-overlapping z-normalised subsequences.

A secondary implicit baseline is the **naïve streaming approach**: recomputing all O(w²) pairwise distances from scratch at every window slide, which has O(w²m) update time.

---

## Limitation of the Baseline Identified by the Paper

The offline algorithm (SDM 2009) operates on a **static, pre-loaded time series**. It requires the entire time series to be available in memory before processing can begin; it cannot accept or incorporate new data points as the stream continues. The paper explicitly states (Section 2.1, "Why is this problem hard?") that when a single data point is inserted or deleted from a window, "the motif pair may have changed, and if it has, then all we know is that the new motif pair has a distance of at most D. The locations of the new motif pair can be anywhere." This means no incremental shortcut exists for the offline algorithm — it must restart from scratch on the new window, which is computationally prohibitive for continuous streams.

The naïve streaming baseline avoids this by recomputing everything from scratch, but at O(w²m) time per update — far too slow for real-time deployment with windows of even a few hundred subsequences.

---

## How the Proposed Method Overcomes the Limitation

The paper introduces a **per-point Neighbour List (N-list) + Reverse List (R-list) data structure** that maintains a pruned, timestamp-ordered set of closest-pair candidates. The two pruning observations (Observation 1: store each pair once; Observation 2: keep only timestamp-monotone, distance-non-increasing N-list entries) together ensure that only the distance computations performed once on insertion are needed — deletions require no distance recomputation at all, only pointer updates. This reduces the worst-case update time to **O(wm) on insertion and O(w) amortised on deletion**, enabling exact, continuously-updated motif answers at stream speed.

---

## Condition Under Which the Paper's Method Would Not Outperform the Baseline

**The baseline offline approach would match or outperform the streaming algorithm in a scenario where the window contents are completely replaced at every update step**—i.e., when w = 1 (a window that holds only one subsequence) or when the stream is "batch-like" and the entire window changes at once. More practically:

> If the stream update rate is so low that the full offline batch algorithm can re-run from scratch before the next data point arrives, the offline approach would produce the same exact answer with no algorithmic overhead of maintaining N-lists and R-lists.

Additionally, the paper itself notes (Section 5, performance experiments) that the streaming algorithm's advantage grows with window size w. For very small windows (e.g., w ≤ 10), the O(w²m) naïve approach is fast enough in absolute time that the constant-factor overhead of maintaining the N-list / R-list data structure may actually make the streaming algorithm slower in practice.

A further reasoned inference: in **multidimensional streams** (Section 6.2 extension), the distance computation is O(m·d) per pair where d is the number of dimensions. For very high dimensional streams (large d), the insertion step's O(wmd) cost dominates and the advantage of the N-list pruning diminishes relative to simply re-running the offline algorithm on the new window.
